```python
import pandas as pd
import seaborn as sns
import re
```

```python
df= pd.read_csv("reviews.csv")
df[:5]
```

```python
df.info()
```

```python
df.text[1]
```

```python
def cleanText(text):
    text= re.sub("[^A-Za-z ]", '', str(text))
    text= text.lower()
    
    return text
```

```python
df["clean_text"]= df.text.apply(cleanText)
df[:5]
```

```python
data= df[['text', 'clean_text','rating', 'rating_category']]
data[:5]
```

```python
data.info()
```

```python
data.text[3]
```

```python
data.clean_text[3]
```

```python
sentiments_analysis= data.clean_text.apply(sentiment_pipeline)
```

```python
sentiments_analysis[:5]
```

```python
type(sentiments_analysis)
```

```python
for label, score in sentiments_analysis.items():
    print(score)
```

```python
#labelScore[['sentiment', 'polarity_score']] = labelScore.label_score.str.extract('([a-zA-Z]+)([0-9]+)')
```

```python
label_score= []

for index, list_ in sentiments_analysis.items():
    for content in list_:
        for values in content.values():
            label_score.append(values)
label_score[:5]
```

```python
polarity_score = []
sentiment = []

for i in label_score:
    if type(i) == str:
        sentiment.append(i)
    else:
        polarity_score.append(i)
        
print(sentiment) 
print(polarity_score)
```

```python
sentiment[:5], polarity_score[:5]
```

```python
sentiment_polarity= pd.DataFrame(list(zip(sentiment, polarity_score)), columns=["sentiment", "polarity_score"])
sentiment_polarity[:5]
```

```python
sentiment_polarity= sentiment_polarity.reset_index().rename(columns= {'index': 'id'})
sentiment_polarity
```

```python
data= data.reset_index().rename(columns={'index': 'id'})
data[:5]
```

```python
sentiment_polarity= sentiment_polarity.merge(data, how= 'left')
sentiment_polarity
```
